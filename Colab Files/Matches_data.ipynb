```python
import pandas as pd
```

```python
df = pd.read_csv('matches.csv')
df.head()
```

```python
# 1 season with most matches
season = df['season'].value_counts()
print("Season with most matches : ",season.idxmax())
```

```python
# 2. Team with most wins
win = df['winner'].value_counts()
print("Team with most wins:", win.idxmax())
```

```python
# 3. Matches in each venue
venue = df['venue'].value_counts()
print("Matches in each venue:\n", venue)
```

```python
# 4. Matches played by each team
teams = df['team1'].value_counts() + df['team2'].value_counts()
print("Matches played by each team:\n", teams)
```

```python
# 5. Wins per team
wins_per_team = df['winner'].value_counts()
print("Wins per team:\n", wins_per_team)
```

```python
# 6. Player of the match
top_players = df['player_of_match'].value_counts()
print("Top players of the match:\n", top_players)
```
